<p>three layers: backend, Artist (does the heavy lifting), scripting (everyday uses)</p>

```python
import numpy as np
# import scipy
# import pandas as pd

from matplotlib.backends.backend_agg import FigureCanvasAgg as FigureCanvas
from matplotlib.figure import Figure

ImportPath = "/home/cerbero/Documents/IBM DV0101EN/00/"
ExportPath = "/home/cerbero/Documents/IBM DV0101EN/01/"
```

```python
fig = Figure()
canvas = FigureCanvas(fig) # creating the objects for matplotlib to plot

x = np.random.randn(10000) #some random numbers

ax = fig.add_subplot(111) #adding an axis to the figure ; 111 is matlab convention for regular x, y axis
```

```python
ax.hist(x, 100) # hist creates the histogram
```

```python
ax.set_title("Normal distribution with $\mu=0, \sigma=1$")
fig.savefig(ExportPath+"matplotlib_histogram.png")
```

```python
from IPython.display import Image
Image(ExportPath+"matplotlib_histogram.png")
```

We can see that the Artist layer example above is somewhat heavy on the syntax. Everyday use of Matplotlib relies on scripting, shown below.

```python
import matplotlib.pyplot as plt   #key to the use of scripts is pyplot, nicknamed as plt

plt.hist(x, 100)
plt.title("Normal distribution with $\mu=0, \sigma=1$")
plt.savefig(ExportPath+"matplotlib_histogram_through_pyplot.png")
plt.show()
```

```python
Image(ExportPath+"matplotlib_histogram_through_pyplot.png")
```

As we just saw, pyplot created the figure, canvas, axis, and so on, all in the background. Much simpler syntax to the same result.
