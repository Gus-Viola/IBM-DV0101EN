```python
fig = Figure()
canvas = FigureCanvas(fig) # creating the objects for matplotlib to plot

x = np.random.randn(10000) #some random numbers

ax = fig.add_subplot(111) #adding an axis to the figure ; 111 is matlab convention for regular x, y axis
```

```python
haiti.index = haiti.index.map(int) # let's change the index values of Haiti to type integer for plotting
haiti.plot(kind='line')

plt.title('Immigration from Haiti')
plt.ylabel('Number of immigrants')
plt.xlabel('Years')

plt.show() # need this line to show the updates made to the figure
```

```python
def PlotImmigrationFrom (country):
    try: 
        listCountry = df_canada.loc[country, years]
        listCountry.plot(kind="line")
        plt.title('Immigration from '+country)
        plt.ylabel('Number of immigrants')
        plt.xlabel('Years')
    except:
        print("I could not loc "+country)
    
#     print(listCountry.head())
```

I coded the generalized function above. It works nicely.
